```python
from torch import nn
class ecg2type(nn.Module):
    def __init__(self,input_size,hidden_size, batch_size, layer=1, output_size=1):
        super(ecg2type,self).__init__()
        self.input_size=input_size
        self.hidden_size=hidden_size
        self.batch_size=batch_size
        self.layer=layer
        self.output_size=output_size
        self.rnn0=nn.RNN(input_size, hidden_size, layer, batch_first=True)
        self.lc=nn.Linear(hidden_size,self.output_size)
    def forward(self,x):
        h0=self.init_hidden(self.layer, self.batch_size, self.hidden_size)
        x,h0=self.rnn0(x,h0)
        res=self.lc(h0.view(self.batch_size,1,self.hidden_size))
        return res.view(self.batch_size,self.output_size)
    def init_hidden(self, layer, batch_size, hidden_size):
        return torch.randn(layer, batch_size, hidden_size)
```
